# Turning our Projections into Fantasy Points

put in all of our previous info

```python
import pandas as pd
player_proj = pd.read_csv('../nba_csv/player_proj_df.csv')
player_info = pd.read_csv('../nba_csv/player_info_df.csv')
```

```python
#slicing and saving to new df 
df_proj_stats = player_proj.loc[:, 
               ['proj_pts','proj_min','proj_fgm','proj_fga','proj_fg3m','proj_fg3a','proj_ftm','proj_fta','proj_oreb','proj_dreb','proj_ast','proj_stl','proj_tov', 'proj_blk']]
#concatenate with player_info to get names, and rename columns to actual stats 
df_proj_stats = pd.concat([player_info, df_proj_stats], axis=1)
df_proj_stats.rename(columns = {'proj_season_id':'season_id',
                               'proj_pts': 'pts',
                                'proj_min':'min',
                                'proj_fgm':'fgm',
                                'proj_fga':'fga',
                                'proj_fg3m':'fg3m',
                                'proj_fg3a':'fg3a',
                                'proj_ftm':'ftm',
                                'proj_fta':'fta',
                                'proj_oreb':'oreb',
                                'proj_dreb':'dreb',
                                'proj_ast':'ast',
                                'proj_stl':'stl',
                                'proj_tov':'tov',
                               'proj_blk':'blk'}, inplace = True)
```

```python
df_proj_stats.sample(5)
```

function that takes row input and output new fantasy point values

using h2h league settings

```python
def fantasy_pts_converter(row):
    pts = row['pts']
    fgm = row['fgm']
    fga = row['fga']
    ftm = row['ftm']
    fta = row['fta']
    oreb = row['oreb']
    dreb = row['dreb']
    ast = row['ast']
    stl = row['stl']
    tov = row['tov']
    blk = row['blk'] 
    fantasy_pts = pts + fgm - fga + ftm - fta + oreb + dreb + ast + stl - tov + blk
    return fantasy_pts
    
```

```python
# now using this method, apply it to the df and create new col for those fantasy pts
fantasy_value = df_proj_stats.apply(fantasy_pts_converter, axis = 1) #saying axis = 1 to apply over each col
#fantasy value saved as a series, so let's put it as a new column in df
df_proj_stats['fantasy_pts'] = fantasy_value
```

```python
df_proj_stats.sort_values('fantasy_pts', ascending = False)
```

now taking care of exporting this as a viable csv file

```python
df_for_export = round(df_proj_stats.sort_values('fantasy_pts', ascending = False),2)
df_for_export.to_csv('../nba_csv/projected_fantasy_pts_final_2018_19.csv', index=False)
```
