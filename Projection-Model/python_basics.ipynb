# Inplace Parameter

### basically, do you want to override the existing df?

### if so (yes), then inplace = True, else inplace = False

### the original df has/has not been modified means inplace = True/False

```python
file_path_updated = '../nba_csv/player_stats_total_updated.csv'
df = pd.read_csv(file_path_updated, encoding='latin1', delimiter=';')
```

```python
df_slim = df.sample(20)
df_slim
```

```python
#modifying this dataframe

df_slim.sort_values('G') 

df_slim
#sorting values by gp, lowest to highest
```

here is what he has annunciated, what I believe is key for understanding inplace parameter:

df_slim.sort_values() is NOT modifying the original df, 

HOWEVER!!!

df_slim = df_slim.sort_values() DOES MODIFY because you are updating memory of df_slim to be the sorted version

NOW, here is what is key. In replacement to the above modifying comes the inplace parameter!!3

```python
df_slim.sort_values('G', inplace = True)

#essentially, no need for "df_slim =" line

```

usually inplace parameter set to false automatically, so when you run "editing" functions, it will run edited version but WONT modify the original DF

We only set Inplace to true if we DO want to overwrite it's information

other types of merges

```python
raw_data = {
        'player_id': ['1', '2', '3', '4', '5', '6', '7'],
        'first_name': ['Lebron', 'Dirk', 'Vince', 'Enes', 'Fred', 'Derrick', 'Paul'], 
        'last_name': ['James', 'Nowitzki', 'Carter', 'Kaner', 'Van Fleet', 'Rose', 'George']}

df_A = pd.DataFrame(raw_data, columns = ['player_id', 'first_name', 'last_name'])

raw_data = {
        'player_id': ['1', '2', '3', '4', '5'],
        'career_ppg': [27.2, 20.7, 17.2, 11.9, 18.8]}

df_B = pd.DataFrame(raw_data, columns = ['player_id', 'career_ppg'])
```

```python
#left merge - keeping left merge cols and attempting to match right col with left

left_df = pd.merge(df_A, df_B, on = 'player_id', how = 'left')

left_df
```

you'll see for any empty spaces right df didnt have that left df did, NaN values assigned

next is right, or right-outer merge/join

left also is left-outer

```python
right_df = pd.merge(df_A, df_B, on = 'player_id', how = 'right')

right_df
```

next is outer, or full outer join. 

Joins all the rows of both df, tries to match but fills ALL empty with NaN

gives decently similar to left-outer

calculating ppg using total gp and total points scored

```python
player_stats = {'player_name': ['Lebron', 'Dirk', 'Wade', 'Curry', 'George'], 
        'season_id': ['2017-18', '2017-18', '2017-18', '2017-18', '2017-18'], 
        'total_points': [2251, 927, 765, 1346, 1734],
        'games_played': [82, 77, 81, 51, 79]}
df = pd.DataFrame(player_stats)

df
```

```python
def get_ppg(row): #takes row as input
    #setting variable to column
    total_points = row[2] #2nd col
    games_played = row[3]
    
    ppg = round((total_points / games_played), 2)
    return ppg
```

```python
df.apply(get_ppg, axis = 1)
```

```python
#saving it to new col

df['ppg'] = df.apply(get_ppg, axis = 1)

df
```

```python
list(df)
```

```python
def ft_calc(row):
    ft_perc = (row['ftm'] / (row['fta'] + .00001))
    return round(ft_perc,2)
```

```python
df['(newly added) ft%'] = df.apply(ft_calc, axis = 1)

df.sample(3)
```

what if we wanted one player's career free throw percentage?

```python
#group by name, sum up values, apply func

final = df.groupby('player_name').sum().apply(ft_calc, axis = 1)

final.sort_values(ascending = False)
```
